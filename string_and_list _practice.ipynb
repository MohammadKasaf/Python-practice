```python
str="python is awesome"

print(f"reverse string is {str[::-1]}")

vowels_count=0
vowels=['a','e','i','o','u']
for i in range(len(str)):
  if str[i] in vowels:
    vowels_count+=1

print(f"number of vowels in string is {vowels_count}")

str_list=str.split(" ")
print(f"number of words in string is {len(str_list)}")

str_without_space=str.replace(" ","")
print(f"string without space is {str_without_space}")
```

```python
duplicates={}

for ch in str:
    if ch==" ":
            continue
    if ch in duplicates:
         value=duplicates.get(ch)
         new_value=value+1
         duplicates[ch]=new_value
    else:
        duplicates[ch]=1
    
for key,value in duplicates.items():
    if value>1:
        print(f" {key} -> {value}")
```

```python
nums = [10, 5, 30, 15, 25]

largest=nums[0]
smallest=nums[0]
second_largest=nums[0]

for num in nums:
    if largest<num:
        largest=num
    if smallest>num:
        smallest=num
    if second_largest<num and num<largest:
        second_largest=num
        
print(f"largest number is {largest}")
print(f"smallest number is {smallest}")
print(f"second largest number is {second_largest}")
```

```python
nums = [1,2,3,2,4,1,5]

nums_set=set()
for num in nums:
    nums_set.add(num)
    
print(f"unique numbers are {list(nums_set)}")
```
