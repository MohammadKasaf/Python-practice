```python
num1 = 24
num2 = 18

salary = 75000
age = 25

number = 24

a = 50
b = 80
c = 30

# Comparison Operator
print("num1 > num2 =", num1 > num2)

# Logical Operators
print("AND Result =", num1 > num2 and salary > 50000)
print("OR Result =", num1 > num2 or age < 18)
print("NOT Result =", not(age < 18))

# Salary Range Check
if salary >= 50000 and salary <= 100000:
    print("Salary is between 50000 and 100000")
else:
    print("Salary is outside the range")

# Positive / Negative / Zero
if number > 0:
    print("Number is Positive")
elif number < 0:
    print("Number is Negative")
else:
    print("Number is Zero")

# Even / Odd
if number % 2 == 0:
    print("Number is Even")
else:
    print("Number is Odd")

# Largest Among Three Numbers
if a >= b and a >= c:
    largest = a
elif b >= a and b >= c:
    largest = b
else:
    largest = c

print("Largest Number =", largest)

# Voting Eligibility
if age >= 18:
    print("Eligible for Voting")
else:
    print("Not Eligible for Voting")
```

```python
text = "Databricks"
factorial_num = 5

print("Numbers from 1 to 10:")
for i in range(1, 11):
    print(i, end=" ")

print("\n")

print("Numbers from 10 to 1:")
for i in range(10, 0, -1):
    print(i, end=" ")

print("\n")

print("Even Numbers from 1 to 100:")
for i in range(2, 101, 2):
    print(i, end=" ")

print("\n")

total = 0

for i in range(1, 101):
    total += i

print("Sum of 1 to 100 =", total)

print()

print("Table of 7:")

for i in range(1, 11):
    print(f"7 x {i} = {7 * i}")

print()

vowels = "aeiouAEIOU"
vowel_count = 0

for ch in text:
    if ch in vowels:
        vowel_count += 1

print("Vowel Count =", vowel_count)

print()

print("Characters in String:")

for ch in text:
    print(ch)

print()

factorial = 1

for i in range(1, factorial_num + 1):
    factorial *= i

print(f"Factorial of {factorial_num} = {factorial}")
```

Print patterns

```python
for i in range(10):
    for j in range(i):
        print("*",end=" ")
    print("\n")

for i in range(10):
    for j in range(10-i):
        print("*",end=" ")
    print("\n")
```

Employee Salary Analysis

```python
employees = [
    {"name": "Amit", "salary": 50000},
    {"name": "Rahul", "salary": 70000},
    {"name": "Neha", "salary": 90000},
    {"name": "Priya", "salary": 120000}
]

print("Employee Names:")

for employee in employees:
    print(employee["name"])

total_salary = 0

for employee in employees:
    total_salary += employee["salary"]

print("\nTotal Salary =", total_salary)

average_salary = total_salary / len(employees)

print("Average Salary =", average_salary)

print("\nEmployees with Salary > 80000:")

for employee in employees:
    if employee["salary"] > 80000:
        print(employee["name"], "-", employee["salary"])

count = 0

for employee in employees:
    if employee["salary"] > 60000:
        count += 1

print("\nEmployees with Salary > 60000 =", count)
```
